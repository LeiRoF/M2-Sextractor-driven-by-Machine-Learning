```python
from numpy import *
from scipy.ndimage.interpolation import rotate
import matplotlib.pyplot as plt

N = 100
E = 2
L = 20

x = N/4
y = N/4

a=random.rand()*360
a_rad = a*pi/180

X, Y = meshgrid(arange(N),arange(N))

pupil = abs(Y - N/2) < E
pupil *= X - N/2 < L
pupil *= 0 < X - N/2
pupil = rotate(pupil, a, reshape=False)
pupil = roll(pupil, (int(x-N/2),int(y-N/2)), axis=(0,1))

plt.pcolor(range(N),range(N),pupil, shading='auto', cmap="gray")
plt.colorbar()
plt.title("Pupil")
plt.show()
```
